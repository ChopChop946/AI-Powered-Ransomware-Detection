# Introduction :
In this notebook, we explore the effectiveness of lightweight machine learning models for ransomware detection using behavioral data. Our objective is to assess how well models like RandomForest, XGBoost, and MLP can distinguish between benign applications and ransomware, both in familiar and unfamiliar scenarios. To achieve this, we begin with a strict family-aware split strategy that simulates the challenge of detecting entirely new ransomware families. This approach reveals the difficulty these models face in generalizing to unseen threats, as performance drops significantly under these conditions. We then relax this constraint and allow the models to train on a mixed dataset, removing the family separation. This shift leads to a dramatic improvement in performance, demonstrating that the models are indeed capable of learning ransomware patterns effectively when data from all families are available during training. Throughout this process, we compare the models under both conditions to better understand their limitations and strengths, ultimately drawing clear conclusions about their practical applications and the importance of generalization in ransomware detection.

```python
import pandas as pd
import numpy as np

df = pd.read_parquet("../Datasets/final_prepared_dataset.parquet")

print("Dataset shape:", df.shape)
print("\nSample rows:")
display(df.head())

print("\nUnique families:", df["family"].unique())

print("\nClass distribution:")
print(df["label"].value_counts())
```

```python
np.random.seed(42)
families = df["family"].unique()
np.random.shuffle(families)

test_families = families[:3]
train_families = families[3:]

print("\nTrain Families:", train_families)
print("Test Families:", test_families)

train_df = df[df["family"].isin(train_families)].reset_index(drop=True)
test_df = df[df["family"].isin(test_families)].reset_index(drop=True)

print("\nTrain set shape:", train_df.shape)
print("Test set shape:", test_df.shape)
```

```python
X = train_df.drop(columns=["label", "family", "window_id"])
y = train_df["label"]
groups = train_df["family"]

folds = {
    1: ["Firefox", "Ryuk", "REvil"],
    2: ["Darkside", "Office", "Zip"],
    3: ["Conti", "LockBit", "Idle"]
}

print("\nManual Folds:")
for fold, val_families in folds.items():
    print(f"Fold {fold} - Val Families:", val_families)

splits = []
for val_families in folds.values():
    val_idx = train_df[train_df["family"].isin(val_families)].index
    train_idx = train_df[~train_df["family"].isin(val_families)].index
    splits.append((train_idx, val_idx))

print("\nPrepared manual splits (by family):")
for fold, (train_idx, val_idx) in enumerate(splits):
    print(f"Fold {fold+1}: Train size {len(train_idx)}, Val size {len(val_idx)}")
```

```python
print("\n=== Manual Folds Overview ===")
for fold_num, (train_idx, val_idx) in enumerate(splits, start=1):
    val_families = folds[fold_num]
    train_families = train_df.iloc[train_idx]["family"].unique()
    
    print(f"\nFold {fold_num}:")
    print(f"â†’ Train Families ({len(train_families)}): {sorted(train_families.tolist())}")
    print(f"â†’ Val Families   ({len(val_families)}): {val_families}")
    print(f"â†’ Train Samples: {len(train_idx)}")
    print(f"â†’ Val Samples  : {len(val_idx)}")
```

```python
from sklearn.metrics import classification_report
import numpy as np

all_y_true = []
all_y_pred = []

for fold_num, (train_idx, val_idx) in enumerate(splits, start=1):
    print(f"\n============================")
    print(f"Fold {fold_num} â†’ Training RandomForest")
    
    X_train = X.iloc[train_idx]
    y_train = y.iloc[train_idx]
    
    X_val = X.iloc[val_idx]
    y_val = y.iloc[val_idx]
    
    rf = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42, n_jobs=-1)
    rf.fit(X_train, y_train)
    
    y_val_pred = rf.predict(X_val)
    
    print(f"\nValidation Report (Fold {fold_num})")
    print(classification_report(y_val, y_val_pred, digits=4))
    
    all_y_true.append(y_val)
    all_y_pred.append(y_val_pred)

all_y_true = np.concatenate(all_y_true)
all_y_pred = np.concatenate(all_y_pred)

print("\n============================")
print("Unified Validation Report (All Folds Combined):")
print(classification_report(all_y_true, all_y_pred, digits=4))
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import classification_report
import numpy as np

all_y_true = []
all_y_pred = []

for fold_num, (train_idx, val_idx) in enumerate(splits, start=1):
    print(f"\n============================")
    print(f"Fold {fold_num} â†’ Training XGBoost")
    
    X_train = X.iloc[train_idx]
    y_train = y.iloc[train_idx]
    
    X_val = X.iloc[val_idx]
    y_val = y.iloc[val_idx]
    
    xgb = XGBClassifier(
        n_estimators=100,
        max_depth=10,
        learning_rate=0.01,
        use_label_encoder=False,
        eval_metric='logloss',
        verbosity=0
    )
    
    xgb.fit(X_train, y_train)
    
    y_val_pred = xgb.predict(X_val)
    
    print(f"\nValidation Report (Fold {fold_num})")
    print(classification_report(y_val, y_val_pred, digits=4))
    
    all_y_true.append(y_val)
    all_y_pred.append(y_val_pred)

all_y_true = np.concatenate(all_y_true)
all_y_pred = np.concatenate(all_y_pred)

print("\n============================")
print("Unified Validation Report (All Folds Combined):")
print(classification_report(all_y_true, all_y_pred, digits=4))
```

```python
from sklearn.neural_network import MLPClassifier

all_y_true = []
all_y_pred = []

for fold_num, (train_idx, val_idx) in enumerate(splits, start=1):
    print(f"\n============================")
    print(f"Fold {fold_num} â†’ Training MLP")
    
    X_train = X.iloc[train_idx]
    y_train = y.iloc[train_idx]
    
    X_val = X.iloc[val_idx]
    y_val = y.iloc[val_idx]
    
    mlp = MLPClassifier(
        hidden_layer_sizes=(256, 128),
        max_iter=200,
        random_state=42,
        early_stopping=True,
        verbose=False
    )
    
    mlp.fit(X_train, y_train)
    
    y_val_pred = mlp.predict(X_val)
    
    print(f"\nValidation Report (Fold {fold_num})")
    print(classification_report(y_val, y_val_pred, digits=4))
    
    all_y_true.append(y_val)
    all_y_pred.append(y_val_pred)

all_y_true = np.concatenate(all_y_true)
all_y_pred = np.concatenate(all_y_pred)

print("\n============================")
print("Unified Validation Report (All Folds Combined):")
print(classification_report(all_y_true, all_y_pred, digits=4))
```

```python
import matplotlib.pyplot as plt

model_names = ["RandomForest", "XGBoost (d10, lr0.1)", "MLP (256,128)"]

accuracies = [0.6867, 0.6929, 0.6261]

benign_f1 = [0.6359, 0.6604, 0.5669]

ransom_f1 = [0.7251, 0.7197, 0.6711]

macro_f1 = [0.6805, 0.6900, 0.6190]

x = np.arange(len(model_names))
width = 0.2

fig, ax = plt.subplots(figsize=(12, 6))
rects1 = ax.bar(x - 1.5*width, accuracies, width, label='Accuracy')
rects2 = ax.bar(x - 0.5*width, benign_f1, width, label='Benign F1')
rects3 = ax.bar(x + 0.5*width, ransom_f1, width, label='Ransomware F1')
rects4 = ax.bar(x + 1.5*width, macro_f1, width, label='Macro F1')

ax.set_ylabel('Scores')
ax.set_title('Validation Performance Comparison')
ax.set_xticks(x)
ax.set_xticklabels(model_names)
ax.legend()

plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
X_test = test_df.drop(columns=["label", "family", "window_id"])
y_test = test_df["label"]

X_train_full = X
y_train_full = y

models = {
    "RandomForest": RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42, n_jobs=-1),
    "XGBoost (d10, lr0.1)": XGBClassifier(n_estimators=100, max_depth=10, learning_rate=0.1, use_label_encoder=False, eval_metric='logloss', verbosity=0),
    "MLP (256,128)": MLPClassifier(hidden_layer_sizes=(256, 128), max_iter=200, random_state=42, early_stopping=True, verbose=False)
}

results = {}

for name, model in models.items():
    print(f"\n============================")
    print(f"{name} â†’ Training on FULL train families")
    
    model.fit(X_train_full, y_train_full)
    
    y_pred = model.predict(X_test)
    
    print(f"\nTest Report ({name})")
    report = classification_report(y_test, y_pred, digits=4, output_dict=True)
    print(classification_report(y_test, y_pred, digits=4))
    
    results[name] = {
        "Accuracy": report["accuracy"],
        "Benign F1": report["0"]["f1-score"],
        "Ransomware F1": report["1"]["f1-score"],
        "Macro F1": report["macro avg"]["f1-score"]
    }
```

```python
model_names = list(results.keys())
accuracies = [results[m]["Accuracy"] for m in model_names]
benign_f1 = [results[m]["Benign F1"] for m in model_names]
ransom_f1 = [results[m]["Ransomware F1"] for m in model_names]
macro_f1 = [results[m]["Macro F1"] for m in model_names]

x = np.arange(len(model_names))
width = 0.2

fig, ax = plt.subplots(figsize=(12, 6))
rects1 = ax.bar(x - 1.5*width, accuracies, width, label='Accuracy')
rects2 = ax.bar(x - 0.5*width, benign_f1, width, label='Benign F1')
rects3 = ax.bar(x + 0.5*width, ransom_f1, width, label='Ransomware F1')
rects4 = ax.bar(x + 1.5*width, macro_f1, width, label='Macro F1')

ax.set_ylabel('Scores')
ax.set_title('Test Split Performance (Unseen Families)')
ax.set_xticks(x)
ax.set_xticklabels(model_names)
ax.legend()

plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## Summary 1:

While the family-aware validation revealed important insights about generalization, the overall performance of the models clearly exposed their limitations in dealing with unseen families. All three models struggled, with accuracy hovering around 50-52% and macro F1 scores reflecting significant difficulty in predicting ransomware correctly when facing new families. Although RandomForest and XGBoost showed slightly better adaptability compared to MLP, even they could not achieve satisfactory generalization. This suggests that, in the current form, the available features and models are not yet capable of fully capturing the deeper patterns necessary to handle completely unseen ransomware behaviors. Given this, a practical decision for now would be to temporarily relax the family-aware constraint and allow the models to train on a mixed dataset. This should help them better exploit the available data, learn more diverse patterns, and achieve higher predictive performance across both ransomware and benign applications, especially when the goal shifts towards creating a robust detector for known threats in realistic environments rather than a purely research-oriented generalization task.

Both RandomForest and XGBoost achieved strong accuracy and balanced F1 scores, with XGBoost reaching over 85% accuracy and demonstrating excellent benign and ransomware detection capabilities. Even MLP, which previously failed to generalize under family-aware conditions, performed significantly better, though it still lagged behind the tree-based models. These results confirm that the models are indeed powerful when the learning task does not involve unseen families. It also reinforces the idea that much of the difficulty observed earlier stemmed from the challenge of generalization rather than weak modeling capacity. Allowing models to learn from a mixed dataset makes them far more capable of capturing diverse behavioral patterns, and thus, for practical purposesâ€”especially when detecting known ransomware in controlled or enterprise environmentsâ€”this training approach may offer a much more effective solution.

# ðŸ“Œ Final Summary

In this notebook, we conducted a thorough and progressive evaluation of multiple lightweight models for the task of ransomware detection using behavioral data. Our process was intentionally designed in stages to better understand both model capabilities and the inherent challenges of generalization, especially when facing unseen ransomware families.
Step 1: Family-Aware Splitting and Validation

We began with a rigorous setup where we divided the dataset based on families. Three families (WannaCry, SDelete, AESCrypt) were held out completely as a test set, while the remaining families were split into train and validation folds in a family-aware manner. This ensured that during validation, models would only be evaluated on families they had not seen during training, simulating the real-world challenge of detecting new ransomware variants.

Three models were tested in this setup:

    RandomForest

    XGBoost (depth 10, learning rate 0.1)

    MLP (256, 128)

The results were sobering.
Model	Accuracy	Macro F1
RandomForest	~49.7%	~49.5%
XGBoost	~52.2%	~51.3%
MLP	~41.5%	~39.6%

All models struggled significantly, particularly on the test split. While RandomForest and XGBoost showed slightly better balance, even they suffered heavily. The models were clearly unable to generalize effectively to ransomware families they had never seen before. This result highlighted the fundamental challenge of unseen-family detection, which requires models to recognize behaviors in patterns that may deviate from those in the training set. In this context, the difficulty was not that the models were weak, but rather that generalization to new ransomware family styles is intrinsically hard given the data and modeling approach.

At this point, it became clear that under strict family-aware constraints, lightweight models alone may not be enough. The models could not confidently and consistently recognize ransomware behaviors without having seen similar family types before.
Step 2: Random Split Without Family Constraints (Mixed Dataset Training)

To better understand the models' raw learning capacity, we performed a second set of experiments where no family constraints were applied. We randomly split the entire dataset into:

    70% for training

    15% for validation

    15% for testing

This allowed the models to train and evaluate on samples from all families, removing the unseen-family generalization problem. The results were dramatically better:
Model	Accuracy	Macro F1
RandomForest	~82.4%	~82.4%
XGBoost	~85.7%	~85.6%
MLP	~66%	~63.9%

This time, both RandomForest and XGBoost demonstrated excellent performance, achieving accuracy levels above 80% and balanced F1 scores for both benign and ransomware samples. Even the MLP, which had previously failed to deliver reasonable results, performed much better in this easier scenario.

These results revealed something crucial: the models are indeed capable of learning and distinguishing ransomware behaviors effectively when the data distribution between training and testing is similar. The large performance gap between the family-aware and random split setups demonstrated that the major challenge was not model capacity, but generalization to unseen ransomware families. When the family-related variation was removed, the models excelled, particularly tree-based ones like XGBoost and RandomForest.
Final Insights and Conclusion

From these experiments, two clear conclusions emerge:

    When the goal is to create a general ransomware detector capable of handling unseen families, family-aware splitting and evaluation are necessary. In this case, lightweight models like RandomForest, XGBoost, and MLP struggle, and alternative methods focusing on better generalization are required. This could include more advanced models, data augmentation, or feature engineering.

    When the goal is to create a ransomware detector for environments where known families are predominant, random splits and mixed dataset training are acceptable and practical. In this scenario, tree-based models, especially XGBoost, show excellent performance and can be deployed effectively.

In practice, the choice of approach depends on the intended deployment context. For open-world detection and research purposes, family-aware evaluation remains mandatory. However, for enterprise security environments where the threat landscape is relatively controlled, training on mixed datasets offers a simpler and highly effective solution.

Overall, this notebook provided valuable insights into both model capabilities and evaluation strategies, setting a clear foundation for further experiments and improvements in ransomware behavioral detection.
