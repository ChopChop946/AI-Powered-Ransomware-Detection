# üìö Ransomware Behavioral Detection ‚Äî LSTM Modeling Summary

This notebook focuses on modeling ransomware detection using **LSTM networks** to better capture the temporal nature of behavioral data.

Each row in our dataset represents a short execution window (0.1 seconds), containing statistical and behavioral features. Ransomware does not reveal itself instantly. Instead, its behavior evolves gradually (increasing entropy, throughput spikes, etc.), which makes sequential modeling essential.

## Dataset Preparation and Sequence Generation

To address this, we generated sequences of consecutive windows:
- Sequences of 20, 30, 40, and 50 windows were tested.
- The label of each sequence was determined by the last window.

This approach gave the model temporal context ‚Üí allowing it to observe how behavior evolved leading up to a window.


# 1 - Load prepared dataset and inspect basic info


```python
import pandas as pd

df = pd.read_parquet("../Datasets/final_prepared_dataset.parquet")

print("Shape of dataset:", df.shape)
print("\nColumn names:\n", df.columns.tolist())
print("\nFirst few rows:")
#display(df.head())

print("\nClass distribution:")
print(df['label'].value_counts())

print("\nFamily distribution:")
print(df['family'].value_counts())
```

# 2 - Build sequential windows for LSTM


```python
import numpy as np
from tqdm import tqdm

SEQUENCE_LENGTH = 40

df_sorted = df.sort_values(by=["family", "window_id"]).reset_index(drop=True)

feature_cols = [col for col in df.columns if col not in ["family", "window_id", "label"]]

sequences = []
sequence_labels = []
sequence_families = []

# Grouping by family to generate sequences independently for each family
for family, group in tqdm(df_sorted.groupby("family")):
    
    group_features = group[feature_cols].values
    group_labels = group["label"].values

    for i in range(len(group) - SEQUENCE_LENGTH + 1):
        seq = group_features[i:i+SEQUENCE_LENGTH]
        label = group_labels[i+SEQUENCE_LENGTH - 1]  # Label of the last window in the sequence
        
        sequences.append(seq)
        sequence_labels.append(label)
        sequence_families.append(family)

sequences = np.array(sequences)
sequence_labels = np.array(sequence_labels)
sequence_families = np.array(sequence_families)

print("Total sequences generated:", sequences.shape[0])
print("Shape of each sequence:", sequences.shape[1:])
print("Sequence labels distribution:")
unique, counts = np.unique(sequence_labels, return_counts=True)
print(dict(zip(unique, counts)))
```

‚ûú Total sequences generated: 194238

    This means that after going through the entire dataset family by family, and sliding over the window_id sorted data, we were able to generate 194,238 sequences of windows.

    Each sequence represents a chunk of execution, containing 20 consecutive windows.

    The number is slightly smaller than the original number of rows (194466) because of how sequences are generated:

    For every family, we generate sequences like this:

- Window 0 ‚Üí Window 19 ‚Üí Sequence 1

- Window 1 ‚Üí Window 20 ‚Üí Sequence 2

- Window 2 ‚Üí Window 21 ‚Üí Sequence 3
- .......  
- Last possible sequence when there are 20 windows left


‚ûú Shape of each sequence: (20, 23)

This is very important:

    20 ‚Üí the sequence length, meaning this sequence contains 20 consecutive windows.

    23 ‚Üí the number of features per window (because we excluded family, window_id, label ‚Üí which means we had 23 features left).

‚ûú Why did we choose "last window" as label?

Because it simulates a very realistic decision-making process:

    "Given what just happened in the last 2 seconds (20 windows), do I raise an alert ‚Üí ransomware OR benign?"

If ransomware process started ‚Üí the entropy, throughput, memory access ‚Üí will start to change gradually ‚Üí and by the last window ‚Üí the LSTM has context to detect this.

# 3 - Preparing Family-Aware and Random Splits


```python
from collections import Counter

family_aware_folds = {
    1: {
        "train_families": ['LockBit', 'WannaCry', 'Office', 'Zip', 'SDelete', 'AESCrypt'],
        "val_families": ['Firefox', 'Ryuk', 'Darkside']
    },
    2: {
        "train_families": ['Conti', 'Firefox', 'Idle', 'LockBit', 'REvil', 'Ryuk'],
        "val_families": ['Darkside', 'Office', 'Zip']
    },
    3: {
        "train_families": ['Darkside', 'Firefox', 'Office', 'REvil', 'Ryuk', 'Zip'],
        "val_families": ['Conti', 'LockBit', 'Idle']
    }
}

FOLD_NUMBER = 1

current_fold = family_aware_folds[FOLD_NUMBER]
train_families = current_fold["train_families"]
val_families = current_fold["val_families"]

test_families = ['Conti', 'Idle', 'REvil']

print(f"üìå Using Fold {FOLD_NUMBER}")
print("Train Families:", train_families)
print("Validation Families:", val_families)
print("Test Families:", test_families)


train_indices = []
val_indices = []
test_indices = []

for idx, family in enumerate(sequence_families):
    if family in train_families:
        train_indices.append(idx)
    elif family in val_families:
        val_indices.append(idx)
    elif family in test_families:
        test_indices.append(idx)

X_family_train = sequences[train_indices]
y_family_train = sequence_labels[train_indices]

X_family_val = sequences[val_indices]
y_family_val = sequence_labels[val_indices]

X_family_test = sequences[test_indices]
y_family_test = sequence_labels[test_indices]



def print_split(name, X, y, families):
    print(f"\n{name} ‚Üí {X.shape[0]} sequences")
    counts = Counter(y)
    print(f"  - Benign: {counts[0]} sequences")
    print(f"  - Ransomware: {counts[1]} sequences")
    print(f"  Families: {families}")

```

```python
print("üìå Family-Aware Split Summary\n")

print("Train (Seen Families):")
print_split("Train", X_family_train, y_family_train, train_families)

print("Validation (Unseen Families):")
print_split("Validation", X_family_val, y_family_val, val_families)

print("Test (Unseen Families):")
print_split("Test", X_family_test, y_family_test, test_families)
```

```python
from sklearn.model_selection import train_test_split
X_random_train, X_random_test, y_random_train, y_random_test = train_test_split(
    sequences, sequence_labels, test_size=0.2, random_state=42, stratify=sequence_labels
)
X_random_train_sub, X_random_val, y_random_train_sub, y_random_val = train_test_split(
    X_random_train, y_random_train, test_size=0.1875, random_state=42, stratify=y_random_train
)
print("üìå Random Split Summary\n")
print_split("Train", X_random_train, y_random_train, np.unique(sequence_families))
print_split("Test", X_random_test, y_random_test, np.unique(sequence_families))
print_split("Validation", X_random_val, y_random_val, np.unique(sequence_families))
```

# 4 - Build and train the baseline LSTM model


```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

sequence_length = X_family_train.shape[1]  # should be 20
num_features = X_family_train.shape[2]     # should be 23

model = Sequential([
    LSTM(256, return_sequences=True, input_shape=(sequence_length, num_features)),
    LSTM(64),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.summary()
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(0.001),
              loss='binary_crossentropy',
              metrics=['accuracy'])

callbacks = [
    EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=10)
]


history = model.fit(
    X_family_train, y_family_train,
    validation_data=(X_family_val, y_family_val),
    epochs=20,
    batch_size=128,
    callbacks=callbacks
)
```

# 5 - Evaluate model performance + confusion matrix + training curves


```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns


def evaluate_model(X, y_true, dataset_name):
    y_pred_probs = model.predict(X).ravel()
    y_pred = (y_pred_probs >= 0.5).astype(int)

    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred)
    rec = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)

    print(f"üìå {dataset_name} Results:")
    print(f"  Accuracy:  {acc:.4f}")
    print(f"  Precision: {prec:.4f}")
    print(f"  Recall:    {rec:.4f}")
    print(f"  F1-Score:  {f1:.4f}\n")

    return y_true, y_pred


val_true, val_pred = evaluate_model(X_family_val, y_family_val, "Validation (Unseen Families)")


test_true, test_pred = evaluate_model(X_family_test, y_family_test, "Test (Unseen Families)")


cm_val = confusion_matrix(val_true, val_pred)
cm_test = confusion_matrix(test_true, test_pred)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.heatmap(cm_val, annot=True, fmt='d', cmap='Blues',
            xticklabels=["Benign", "Ransomware"],
            yticklabels=["Benign", "Ransomware"],
            ax=axes[0])
axes[0].set_xlabel("Predicted")
axes[0].set_ylabel("Actual")
axes[0].set_title("Confusion Matrix - Validation (Unseen Families)")

sns.heatmap(cm_test, annot=True, fmt='d', cmap='Oranges',
            xticklabels=["Benign", "Ransomware"],
            yticklabels=["Benign", "Ransomware"],
            ax=axes[1])
axes[1].set_xlabel("Predicted")
axes[1].set_ylabel("Actual")
axes[1].set_title("Confusion Matrix - Test (Unseen Families)")

plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)  
plt.plot(history.history['loss'], label="Train Loss", linewidth=2)
plt.plot(history.history['val_loss'], label="Validation Loss", linewidth=2)
plt.title("Loss (Family-aware Split)")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.ylim(0, 5)
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)  
plt.plot(history.history['accuracy'], label="Train Accuracy", linewidth=2)
plt.plot(history.history['val_accuracy'], label="Validation Accuracy", linewidth=2)
plt.title("Accuracy (Family-aware Split)")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.ylim(0, 1)
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

# 6 - Build and train LSTM on Random Split


```python
random_model = Sequential([
    LSTM(128, return_sequences=True, input_shape=(sequence_length, num_features)),
    LSTM(64),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

random_model.compile(optimizer=tf.keras.optimizers.Adam(0.001),
                     loss='binary_crossentropy',
                     metrics=['accuracy'])

callbacks = [
    EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2)
]


history_random = random_model.fit(
    X_random_train_sub, y_random_train_sub,
    validation_data=(X_random_val, y_random_val),
    epochs=20,
    batch_size=128,
    callbacks=callbacks
)
```

```python

def evaluate_model(X, y_true, dataset_name, model):
    y_pred_probs = model.predict(X).ravel()
    y_pred = (y_pred_probs >= 0.5).astype(int)

    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred)
    rec = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)

    print(f"üìå {dataset_name} Results:")
    print(f"  Accuracy:  {acc:.4f}")
    print(f"  Precision: {prec:.4f}")
    print(f"  Recall:    {rec:.4f}")
    print(f"  F1-Score:  {f1:.4f}\n")

    return y_true, y_pred



test_true, test_pred = evaluate_model(X_random_test, y_random_test, "Random Split Test Set", random_model)


def plot_conf_matrix(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6,5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=["Benign", "Ransomware"], yticklabels=["Benign", "Ransomware"])
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title(title)
    plt.show()

plot_conf_matrix(test_true, test_pred, "Confusion Matrix - Random Split Test Set")


```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)  
plt.plot(history_random.history['loss'], label="Train Loss", linewidth=2)
plt.plot(history_random.history['val_loss'], label="Validation Loss", linewidth=2)
plt.title("Loss (Random Split)")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.ylim(0, 0.5)  
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)  
plt.plot(history_random.history['accuracy'], label="Train Accuracy", linewidth=2)
plt.plot(history_random.history['val_accuracy'], label="Validation Accuracy", linewidth=2)
plt.title("Accuracy (Random Split)")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.ylim(0, 1) 
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

# üìå Final Summary of LSTM Notebook and Experiments

This notebook marked an important phase in our ransomware behavioral detection project.
We moved past CNNs and stepped into the sequential modeling territory with LSTM, which was more suited for this task.
Here‚Äôs how everything unfolded step by step.
Dataset Preparation and Splitting Logic

The dataset we worked with consists of short execution windows (0.1 seconds each) enriched with behavioral statistics (entropy, throughput, memory access patterns...).
Each row represents a small "slice of time" and is labeled as benign or ransomware.

However, a single window does not tell the full story of ransomware behavior.
Ransomware typically evolves during execution ‚Üí It starts calm and progressively increases entropy, throughput, and aggressive IO patterns.
Because of this, we decided to shift our modeling paradigm from classifying isolated windows ‚Üí to classifying sequences of windows.
Sequence Generation

To give the model temporal context, we generated sequences using a sliding window approach.
Each sequence consisted of 20, 30, 40, or even 50 consecutive windows.
The label of each sequence was defined by the label of its last window ‚Üí because that reflects the "current state" of the process when the sequence ends.
Split Strategies

Understanding ransomware is not only about accuracy ‚Äî it‚Äôs about robustness.
So, we followed two evaluation protocols:
## 1Ô∏è‚É£ Random Split

    Dataset randomly split into train, validation, and test sets.

    All families mixed.

    Easier setup ‚Üí model sees patterns from every family during training ‚Üí almost perfect test scores.

    This split is helpful to see if the model can learn behaviors, but very misleading when it comes to true ransomware detection.

    In small datasets (like ours with 12 families), this approach tends to exaggerate performance.

## 2Ô∏è‚É£ Family-Aware Split (The Real Challenge)

    Much stricter ‚Üí holdout families entirely for validation and testing.

    Train on a subset of families ‚Üí test if the model can detect ransomware from families it never saw before.

    This simulates the real-world challenge ‚Üí ransomware authors constantly produce new variants.

For this, we defined folds where train, validation, and test families were carefully selected to maximize diversity.
LSTM Architectures and Experiments

Once sequences were generated and splits prepared, we experimented extensively with LSTM architectures:

    We tested different architectures:

        256 ‚Üí 128 ‚Üí 64 ‚Üí Dense

        256 ‚Üí 128 ‚Üí 32 ‚Üí Dense

        256 ‚Üí 64 ‚Üí 32 ‚Üí Dense (most successful)

    And different sequence lengths:

        20 ‚Üí decent performance but limited generalization

        30 ‚Üí better temporal context ‚Üí stronger generalization

        40 ‚Üí optimal ‚Üí best balance between having enough context without making sequences too long and noisy

        50 ‚Üí still good but slightly degraded, probably due to sequences being too long and containing redundant or diluted information.

## Optimal Model

‚Üí The clear winner across experiments was:

LSTM(256) ‚Üí LSTM(64) ‚Üí Dense(32) with Sequence length = 40.

    On family-aware split:

        ~85% accuracy

        ~88-89% F1 score

        ~82% recall ‚Üí showing strong ability to detect unseen ransomware families

This model struck the best trade-off between generalization and overfitting.
Why LSTM Helped, but Also Its Limitations

LSTM dramatically improved over CNN because it could leverage sequential information ‚Üí ransomware behaviors are patterns across time, not isolated moments.

However, it still has limitations:

    LSTM looks only at the sequence you give it.
    If too short ‚Üí can‚Äôt understand ransomware evolution properly.
    If too long ‚Üí starts forgetting the start or overfitting to noise ‚Üí performance degrades.

    Ransomware is highly diverse.
    Some families encrypt aggressively ‚Üí others stealthily. LSTM alone struggles to create universal representations.

    Family-aware test performance is still far from random split ‚Üí generalization is very hard.
